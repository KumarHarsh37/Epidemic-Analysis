```python
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt

# Load the data
data = ts_Andaman_clean

# Filter and process the data for time series prediction
data = ts_Andaman_clean[['Date', 'Positive']]  # Adjust column names if needed
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Scale the data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data['Positive'].values.reshape(-1, 1))

# Prepare the dataset for LSTM
def prepare_data(series, n_steps):
    X, y = [], []
    for i in range(len(series) - n_steps):
        X.append(series[i:i+n_steps])
        y.append(series[i+n_steps])
    return np.array(X), np.array(y)

# Set the number of time steps
n_steps = 30
X, y = prepare_data(scaled_data, n_steps)

# Reshape X to be compatible with LSTM
X = X.reshape((X.shape[0], X.shape[1], 1))

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, activation='relu', return_sequences=True, input_shape=(n_steps, 1)))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')

# Train the model
model.fit(X, y, epochs=50, verbose=1)

# Forecast future values
n_forecast = 30
forecast_input = scaled_data[-n_steps:]
forecast = []

for _ in range(n_forecast):
    forecast_input = forecast_input.reshape((1, n_steps, 1))
    prediction = model.predict(forecast_input, verbose=0)
    forecast.append(prediction[0, 0])
    forecast_input = np.append(forecast_input[0, 1:], prediction[0, 0])

# Inverse scale the forecast
forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

# Plot the results
plt.plot(data.index, data['Positive'], label="Actual")
future_dates = pd.date_range(data.index[-1], periods=n_forecast+1, freq='D')[1:]
plt.plot(future_dates, forecast, label="Forecast")
plt.xlabel("Date")
plt.ylabel("Positive Cases")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Plot the actual data
plt.plot(data.index, data['Positive'], label="Actual", color='blue')

# Plot the forecasted data
future_dates = pd.date_range(data.index[-1], periods=n_forecast+1, freq='D')[1:]
plt.plot(future_dates, forecast, label="Forecast", color='red', linestyle='--')

# Set labels and title
plt.title('Actual vs Forecasted Positive Cases')
plt.xlabel("Date")
plt.ylabel("Positive Cases")
plt.legend()
plt.grid(True)

plt.show()
```

```python
train_size = int(len(scaled_data) * 0.8)
train_data = scaled_data[:train_size]
test_data = scaled_data[train_size - n_steps:]  # Include a buffer for lag

# Prepare training and test data
X_train, y_train = prepare_data(train_data, n_steps)
X_test, y_test = prepare_data(test_data, n_steps)

# Reshape X to match LSTM input format
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
model.fit(X_train, y_train, epochs=50, verbose=1)
```

```python
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred)  # Inverse scale predictions
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))  # Inverse scale actual values
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
import numpy as np

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100  # MAPE in percentage

# Print metrics
print(f'Mean Squared Error (MSE): {mse}')
print(f'Root Mean Squared Error (RMSE): {rmse}')
print(f'Mean Absolute Error (MAE): {mae}')
print(f'Mean Absolute Percentage Error (MAPE): {mape:.2f}%')
```

```python
# Calculate MAPE
mape = mean_absolute_percentage_error(y_test, y_pred) * 100  # MAPE in percentage

# Calculate accuracy as (100% - MAPE)
accuracy = 100 - mape

# Print accuracy
print(f'Model Accuracy: {accuracy:.2f}%')
```
